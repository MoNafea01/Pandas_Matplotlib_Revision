#### Part 9 | Cleaning Data

```python
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
df2 = pd.DataFrame(people)
df2
```

```python
df2.dropna(axis='rows',how='all',inplace=True) # axis= 'rows' drops the rows which have NaN values, how='any' drops the entire row because it have at least
#one missing value , if you changed it to how='all' then it will drop the row if all columns have missing values
df2
```

```python
df2.dropna(axis='rows',how='any',subset=['email'],inplace=True) # subset used to select a specific column and apply other arguments to this column
# so if any row have a null value in the email section it will be dropped (note: NA isn't a null object it is just a string)
df2
```

```python
df2.replace({'NA':np.nan,'Missing':np.nan},inplace=True)
df2.dropna(axis=0,how='all',subset=['email','last'],inplace=True)
df2
```

```python
df2.isna() # used to ask each value if it is a null or not
```

```python
df2.isna().any() # ask the columns if it has any null value 
```

```python
df2['age'] = df2['age'].astype(float) # why float and not int (if you used int you will get an error) the cause of that is NaN is actually a float
# number so you can represent it as float and that's why an error occurs when you use int
df2.dtypes
```

```python
df2.fillna(value= {'age':pd.Series(data=df2['age'].mean(),index=[0,1,2,5])})
```
